## importing libraries and reading data

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
athlete_df = pd.read_csv('data/olympics/athlete_events.csv')
region_df = pd.read_csv('data/olympics/noc_regions.csv')
```

```python
goldMedals = data[(data.Medal == 'Gold')]
# to take values that is different from nan 
goldMedals = goldMedals[np.isfinite(goldMedals['Age'])]
```

```python
# gold medal distribution
plt.figure(figsize=(26, 18))
plt.tight_layout()
sns.countplot(goldMedals['Age'])
plt.title('Distribution of Gold Medals')
```

Age 23 and 24 are the most ages that acheive gold medals

```python
masters = goldMedals['Sport'][goldMedals['Age'] > 50]
```

```python
plt.figure(figsize=(20, 10))
plt.tight_layout()
sns.countplot(masters)
plt.title('Gold Medals for Athletes Over 50')
```

```python
# top 20 couontries participating
top_10_countries = data.Team.value_counts().sort_values(ascending = False).head(20)
plt.style.use('fivethirtyeight')
fig = plt.figure(figsize=(18,8))
plt.title('Top 10 countries participate in the olympicss',size = 25)
top_10_countries.plot(kind = 'barh')
```

```python
# top 10 countries with gold medals
gold_medals = data.query('Medal == "Gold"')
total_gold_medals = gold_medals.Region.value_counts().head(15)
plt.style.use('fivethirtyeight')
fig = plt.figure(figsize=(18,8))
total_gold_medals.plot(kind = 'bar')
plt.title('Top 10 countries with most gold medals',size = 25)
plt.xlabel('gold medals')
```

The countries with the most participation are the countries with the most gold medals

france participate more but not acheived a lot of medals

## Women in Athletics

```python
women_in_olympics = data[(data['Sex'] == 'F') & (data.Season == 'Summer')]
women_in_olympics.head(10)
```

```python
sns.set(style="darkgrid")
plt.figure(figsize=(20, 10))
sns.countplot(x='Year', data=women_in_olympics)
plt.title('Women medals per edition of the Games')
```

```python
MenOverTime = data[(data.Sex == 'M') & (data.Season == 'Summer')]
sns.set(style="darkgrid")
plt.figure(figsize=(20, 10))
part = MenOverTime.groupby('Year')['Sex'].value_counts()
part.loc[:,'M'].plot()
plt.title('men participation over time')
```

```python
sns.set(style="darkgrid")
plt.figure(figsize=(20, 10))
part = women_in_olympics.groupby('Year')['Sex'].value_counts()
part.loc[:,'F'].plot()
plt.title('Women participation over time')
```

We have a steep increase in the population

The grow is constant.

On the other hand, the grow for men seems less strong:

After the 1990 we can see a relevant decrease in the number of male athletes at the summer games

The growth has slowly restarted recently.
